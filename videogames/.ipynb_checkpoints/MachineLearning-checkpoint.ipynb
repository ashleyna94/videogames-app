```python
# Import necessary libraries
import os
import numpy as np
import pandas as pd
from datetime import datetime
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from flask import (
    Flask, render_template, jsonify,
    request, redirect)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Database Setup
engine = create_engine("sqlite:///video_games.sqlite")

Base = automap_base()
Base.prepare(engine, reflect=True)

Videogames = Base.classes.videogames
session = Session(engine)
```

```python
# Read into Pandas
stmt = session.query(Videogames).statement
df = pd.read_sql_query(stmt, session.bind)
```

```python
df.head()
```

```python
df = df[df["Critic_Score"]!=" "]
df
```

```python
df.dtypes
```

```python
# Converting columns to numbers
df_nums = df.apply(pd.to_numeric, errors="ignore")
df_nums.head()
```

```python
df_nums.dtypes
```

```python
df_nums2 = df_nums.dropna(axis=0, how="any")
df_nums2.head()
```

```python
# Visualizing Global Sales vs Critic Score
sns.set_style("whitegrid")
fig = plt.subplots(figsize=(10,5))
sns.regplot(x="Critic_Score", y="Global_Sales", data=df_nums2, color="#3bb44b").set(ylim=(0, 37))
plt.show()
```

# Prediction

```python
predict_df = df_nums2.copy()
predict_df.head()
```

```python
predict_df.drop(["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Critic_Count", "User_Score", "User_Count", "Rating"], axis=1, inplace=True)
```

```python
predict_df.head()
```

```python
predict_df["Hit"] = predict_df["Global_Sales"]
predict_df.head(1)
```

```python
def recode(series):
    if series >= 1:
        return 1
    else:
        return 0

predict_df["Hit"] = predict_df["Hit"].apply(recode)
predict_df.head()
```

```python
predict_test = predict_df.copy()
predict_test = predict_test[["Year_of_Release", "Critic_Score", "Hit"]]
```

```python
X = predict_test.drop("Hit", axis=1)
y = predict_test["Hit"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True).head()
```
